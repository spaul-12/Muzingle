```python
#function to create One hot encoding features
def one_hot_encoding(df,column, new_name):
    tf_df = pd.get_dummies(df[column])
    feature_name= tf_df.columns #list of features
    tf_df.columns = [new_name + "|" +str(i) for i in feature_names]
    tf_df.reset_index(drop=True, inplace =True)
    return tf_df
```

```python
def feature_set(df, float_cols):
    
    #tfidf unique genre list
    tfidf=TfidfVectorizer()
    tfidf_Matrix=tfidf.fit_transform(df['unique_genre_list'].apply(lambda x: " ".join(x)))
    genre_df=pd.Dataframe(tfidf_Matrix.toarray())
    genre_df.columns=['genre'+"|" + i for i in tfidf.get_features_names()]
    genre_df.reset_index(drop=True, inplace=True)
    
    #one hot encode year_bucket column
    year_ohe=one_hot_encoding(df,'year','year')*0.5
    popularity_ohe=one_hot_encoding(df,one_hot_encoded_cols,'pop')
    
    #scaling float columns
    floats = df[float_cols].reset_index(drop=True)
    scaler=MinMaxScaler()
    floats_scaled=pd.Dataframe(scaler.fit_transform(floats),columns=floats.columns)*0.2
    
    final_df=pd.concat([genre_df,floats_scaled,popularity_ohe,year_ohe],axis='columns')
    
    final_df['id']=df['id'].values
    
    return final_df
    
    
```
